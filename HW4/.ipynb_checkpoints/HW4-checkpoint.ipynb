```python
import urllib.request as req
import json
import pandas as pd
import numpy as np

url= "https://www.ptt.cc/bbs/KoreanPop/M.1616662853.A.67B.html"
request=req.Request(url, headers={
    "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/107.0.0.0 Safari/537.36"
})

with req.urlopen(request) as response:
    data=response.read().decode("utf-8")
    
```

```python
import bs4
root=bs4.BeautifulSoup(data, "html.parser")
articles = root.find_all('div', 'push')
for article in articles:
    messages = article.find('span', 'f3 push-content').getText()
    
temp = []
for article in articles:
    messages = article.find('span', 'f3 push-content').getText()
    temp.append(messages)
    
df = pd.DataFrame(temp)
df
```

```python
#文字探勘
import jieba
import jieba.analyse
import pandas as pd
import numpy as np
from collections import Counter
```

```python
# 測試結巴功能
documents = ["我的偶像是IU"]
# 精確模式
for sentence in documents:
    seg_list = jieba.cut(sentence)
    print('/'.join(seg_list))
```

```python
df['seg'] = " "
index = 0
for documents in df[0]:
    seg_list = jieba.analyse.extract_tags(documents, 150)
    seg_list = jieba.lcut(documents)
    df['seg'].iloc[index] = seg_list
    index += 1

df['seg']
```

```python
#合成dict
DICT = df['seg'].to_list()
print(DICT)
```

```python
#將dict統整成list
List = []
for i in range(0,210):
  for value in DICT[i]:
    List.append(value)
```

```python
valueToBeRemoved = '啊' 

List = [value for value in List if value != valueToBeRemoved]

results = pd.value_counts(List)
results
```

```python
#用來產生文字雲
from PIL import Image
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
stopwords = {"","推",'\n'}
wc = WordCloud(font_path= "C:\Windows\Fonts\msjh.ttc", 
               background_color="white", 
               max_words = 200 ,       
               stopwords=stopwords)      

wc.generate_from_frequencies(results)
 
# 視覺化呈現
plt.imshow(wc)
plt.axis("off")
plt.figure(figsize=(20,10), dpi =20)
plt.show()
```
